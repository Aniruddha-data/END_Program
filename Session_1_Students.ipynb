```
with open("google-10000-english-no-swears.txt") as f:
    passage = f.read()
    words = passage.lower().split('\n')

print(f'No of words: {len(words)}')
print(f'Some of the words: {words[460:470]}')
```

```
one_hot_vec = torch.zeros(len(words), len(words))
```

```
# Encode rank as embeddings

embeds = torch.nn.Embedding(len(words), 128)
rank_tensor = torch.LongTensor(rank_list)
embeddings = embeds(rank_tensor)

print(f'One hot vector size: {embeddings.shape}')
```
